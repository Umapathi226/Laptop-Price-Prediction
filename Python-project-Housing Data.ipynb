```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
Housingdata=pd.read_excel("Housing data.xlsx")
```

```python
Housingdata.head()
```

```python
# compare avg bed rooms by totalrooms
# Bar chart

plt.figure(figsize=(8,6))
sns.barplot(x='total_rooms',y='total_bedrooms',data=Housingdata,estimator="mean",color='blue')
plt.title('Avg Bed rooms by totalrooms')
plt.xlabel('total_rooms')
plt.ylabel('total_bedrooms')
plt.show()
```

```python
avg_bedrooms=Housingdata.groupby('total_rooms')['total_bedrooms'].mean()
avg_bedrooms
```

```python
# distribution-
# scatter plot- relation ship between two numeric values

plt.figure(figsize=(8,6))
sns.scatterplot(x='total_rooms',y='total_bedrooms',data=Housingdata,color='blue',edgecolor='k')
plt.title('total_rooms vs total_bedrooms')
plt.xlabel('total_rooms')
plt.ylabel('total_bedrooms')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

Housing_data= pd.read_excel("Housing_data.xlsx")

# Bar plot: Housing Median Age vs Median House Value

plt.figure(figsize=(8, 6))
sns.barplot(x='housing_median_age', y='median_house_value', data= Housing_data,color='blue')
plt.title('Housing Median Age vs Median House Value',fontsize=16)
plt.xlabel('Housing Median Age',fontsize= 12)
plt.ylabel('Median House Value',fontsize=12)
plt.show()
```

```python
# Box plot for House Value by Ocean Proximity
plt.figure(figsize=(8, 6))
sns.boxplot(x='ocean_proximity', y='median_house_value', data=Housing_data,color='blue')
plt.title('House Value by Ocean Proximity',fontsize=10)
plt.xlabel('Ocean Proximity',fontsize=8)
plt.ylabel('Median House Value',fontsize=8)
plt.show()
```

```python
# Bar plot - comparision
# house value for ocean proximity

plt.figure(figsize=(6, 4))
sns.barplot(x='ocean_proximity', y='median_house_value', data=Housing_data,color='blue',edgecolor='r')
plt.title('House Value by Ocean Proximity',fontsize=10)
plt.xlabel('Ocean Proximity',fontsize=8)
plt.ylabel('Median House Value',fontsize=8)
plt.show()
```
